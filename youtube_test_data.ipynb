```python
#                                                    STEP - 3 : [ Get Channels Information]  
#               
import pandas as pd              

# USING FOR LOOP [get any particular data from response [by using keys]]

# create a function to pass the channel id aand get channels details

def get_channels_info(channel_id):

    # Get Channels informations

    # inside the list() => pass the parameters 
    request = youtube.channels().list(
    
      # pass parameters : snippet , ContentDetails and statistics 
      
      part = "snippet, ContentDetails, statistics",

      # give youtube channel Id
      
      #id = "UCJcCB-QYPIBcbKcBQOTwhiA"    (VJ SIDDHU VLOGS)
      
      id = channel_id
    )

    response = request.execute()

    channel_data = []
    
    for i in response['items']:
    
      # create a variable and get the particular data based on needs and store it to the variable in dictionary format.
    
      get_data_from_response = dict(Channel_Name = i['snippet']['title'],
                                    Channel_Id = i['id'],
                                    Subscription_Count = i['statistics']['subscriberCount'],
                                    Channel_Views = i['statistics']['viewCount'],
                                    Channel_Description = i['snippet']['description'], # this channel has no description.so 'null'
                                    Playlist_Id = i['contentDetails']['relatedPlaylists']['uploads'] # ~= uploads
                              )
    
      channel_data.append(get_data_from_response)
    
    channel_data_df = pd.DataFrame(channel_data)

    return channel_data_df

# Get_channel_Details = get_channels_info("UCMyiKJ5PO2TIJhQkZqgpffw")    # give any 'channel_id' to get the  channel details 

Get_channel_Details = get_channels_info("UCJcCB-QYPIBcbKcBQOTwhiA")    # give any channel id 

Get_channel_Details


    

```

```python
#                                  Step 4 : [ Get all videos  [Using function] => give channel Id as argument to collect all video Ids ]

#get video ids

def get_all_video_ids(channel_id):

    video_ids = []

    get_playlist_id = youtube.channels().list(id=channel_id,
                                    part='contentDetails').execute()

    Playlist_Id = get_playlist_id['items'][0]['contentDetails']['relatedPlaylists']['uploads']

    next_page_token = None

    while True:

        get_video_id = youtube.playlistItems().list(
                                            part='snippet',
                                            playlistId=Playlist_Id,
                                            maxResults=50,
                                            pageToken=next_page_token).execute()

        for i in range(len(get_video_id['items'])):

            video_ids.append(get_video_id['items'][i]['snippet']['resourceId']['videoId'])

        next_page_token = get_video_id.get('nextPageToken')

        if next_page_token is None:

            break
            
    return video_ids
```

```python
# Calling the get_all_video_ids Function ....to return the video_ids

videoIDs = get_all_video_ids("UCJcCB-QYPIBcbKcBQOTwhiA")   # give any channel ID

videoIDs
```

```python
#    STEP : 7    [Get Playlist Details]

# By using function : passing the channel Id to return the all_playlist data


def get_playlist_details(channel_id):

        next_page_token = None

        playlist_name = []

        while True:

                #  API REFERENCE LINK :               https://developers.google.com/youtube/v3/docs/playlists/list?apix=true

                request = youtube.playlists().list(

                        part = 'snippet,contentDetails',
                        channelId = channel_id,
                        maxResults = 50,
                        pageToken = next_page_token
                )

                response = request.execute()

                for i in response['items']:
                    
                        playlist_data = dict(Playlist_Id = i['id'],
                                             Channel_Id = i['snippet']['channelId'],
                                             Playlist_Name = i['snippet']['title'],
                                             #  Channel_Name = i['snippet']['channelTitle'],
                                             Video_Count = i['contentDetails']['itemCount'],
                                             PublishedAt = i['snippet']['publishedAt'])

                        playlist_name.append(playlist_data)

                next_page_token=response.get('nextPageToken')

                if next_page_token is None:

                        break
        return playlist_name


# Display the results of  playlist details

playlist_info = get_playlist_details("UCJcCB-QYPIBcbKcBQOTwhiA")    # channel_id for VJ SIDDHU channel

playlist_info

```
